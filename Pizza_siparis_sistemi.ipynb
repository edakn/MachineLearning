```python
#Gerekli kütüphaneleri ekledim.
import csv
import datetime
import re
```

```python
#Menu.txt adlı bir menü dosyası oluşturdum ve içine menü metnini yazdım.
with open("Menu.txt", "w") as file:
    file.write("* Lütfen Bir Pizza Tabanı Seçiniz:\n1: Klasik\n2: Margarita\n3: Türk Pizza\n4: Sade Pizza\n\n* ve seçeceğiniz sos:\n11: Zeytin\n12: Mantarlar\n13: Keçi Peyniri\n14: Et\n15: Soğan\n16: Mısır\n\n* Teşekkür ederiz!")
```

```python
#Bir Pizza üst sınıfı oluşturarak, belirli pizza türleri bu ortak özellikleri ve yöntemleri devralıp,
#ardından sınıf özelliklerini ve davranışlarını bunlara ek olarak tanımladım.
class Pizza:
    def __init__(self):
        self._description = "Pizza"
        self._cost = 0.0
    
    def get_description(self):
        return self._description
    
    def get_cost(self):
        return self._cost
```

```python
#Her pizzanın kendine ait bir fiyatı ve sahip olduğu değişkenin içinde pizzaların açıklamasının yer aldığı alt sınıfları tanımladım.
class KlasikPizza(Pizza):
    def __init__(self):
        super().__init__()
        self._description = "Klasik Pizza"
        self._cost = 9.99
    
class MargeritaPizza(Pizza):
    def __init__(self):
        super().__init__()
        self._description = "Margerita Pizza"
        self._cost = 11.99

class TurkPizza(Pizza):
    def __init__(self):
        super().__init__()
        self._description = "Türk Pizza"
        self._cost = 12.99
    
class SadePizza(Pizza):
    def __init__(self):
        super().__init__()
        self._description = "Sade Pizza"
        self._cost = 10.99
```

```python
#Dekoratör sınıfını, çalışma zamanında bir Pizza nesnesine Pizza nesnesinin davranışını etkilemeden ek davranış eklemek için Sos sınıfında kullanmak için oluşturdum.
class Decorator(Pizza):
    def __init__(self, pizza):
        self._pizza = pizza
    
    def get_description(self):
        return self._pizza.get_description()
    
    def get_cost(self):
        return self._pizza.get_cost()
```

```python
#Oluşturduğum sos sınıflarında, bir Pizza nesnesini bağımsız değişken olarak alıyorum ve seçilen sosun açıklamasını ve maliyetini 
#pizzanın mevcut açıklamasına ve maliyetine ekleyerek onu dekore ediyorum.
class Zeytin(Decorator):
    def __init__(self, pizza):
        super().__init__(pizza)
        self._description = "Zeytin"
        self._cost = 1.25
    
    def get_description(self):
        return self._pizza.get_description() + ", " + self._description
    
    def get_cost(self):
        return self._pizza.get_cost() + self._cost

class Mantar(Decorator):
    def __init__(self, pizza):
        super().__init__(pizza)
        self._description = "Mantarlar"
        self._cost = 2.5
    
    def get_description(self):
        return self._pizza.get_description() + ", " + self._description
    
    def get_cost(self):
        return self._pizza.get_cost() + self._cost

class KeciPeyniri(Decorator):
    def __init__(self, pizza):
        super().__init__(pizza)
        self._description = "Keçi Peyniri"
        self._cost = 3.99
    
    def get_description(self):
        return self._pizza.get_description() + ", " + self._description
    
    def get_cost(self):
        return self._pizza.get_cost() + self._cost

class Et(Decorator):
    def __init__(self, pizza):
        super().__init__(pizza)
        self._description = "Et"
        self._cost = 6.99
    
    def get_description(self):
        return self._pizza.get_description() + ", " + self._description
    
    def get_cost(self):
        return self._pizza.get_cost() + self._cost

class Sogan(Decorator):
    def __init__(self, pizza):
        super().__init__(pizza)
        self._description = "Soğan"
        self._cost = 1.5
    
    def get_description(self):
        return self._pizza.get_description() + ", " + self._description
    
    def get_cost(self):
        return self._pizza.get_cost() + self._cost

class Mısır(Decorator):
    def __init__(self, pizza):
        super().__init__(pizza)
        self._description = "Mısır"
        self._cost = 2.5
    
    def get_description(self):
        return self._pizza.get_description() + ", " + self._description
    
    def get_cost(self):
        return self._pizza.get_cost() + self._cost
```

```python
# Pizza ve sos seçeneklerini kullanacağım menüye göre tanımladım.
PIZZA = {
    "1": KlasikPizza, "2": MargeritaPizza,"3": TurkPizza,"4": SadePizza
}

SOS = {
    "11": Zeytin, "12": Mantar,"13": KeciPeyniri,"14": Et,"15": Sogan,"16": Mısır
}

def main():
    # Önce oluşturduğum menüyü yazdırıyorum.
    with open("Menu.txt", "r") as file:
        menu_text = file.read()

    # Menü metnini yıldızlara göre ayırdım.
    menu_parts = menu_text.split("*")

   # Menünün ikinci ve üçüncü bölümlerinden pizza seçeneklerini aldım.
    pizza_options = menu_parts[1].strip()
    print(pizza_options)

    # Kullanıcının pizza ve sos seçeneklerini aldım, herhangi bir hatalı giriş için hata uyarısı verdirdim.
    pizza_secimi = None
    while pizza_secimi not in PIZZA.keys():
          pizza_secimi = input("\nLütfen sipariş etmek istediğiniz pizzanın numarasını giriniz (1-4): ")
          if pizza_secimi not in PIZZA.keys():
             print("Hatalı Pizza numarası girdiniz, lütfen geçerli bir numara girin.")
    
    #Menünün birinci ve üçüncü bölümlerinden sos seçeneklerini aldım.
    sos_options = menu_parts[2].strip()
    print("\n"+sos_options)

    sos_secimi = None
    while sos_secimi not in SOS.keys():
          sos_secimi = input("\nLütfen eklemek istediğiniz sosun numarasını giriniz (11-16): ")
          if sos_secimi not in SOS.keys():
             print("Hatalı bir sos numarası girdiniz, lütfen geçerli bir numara girin.")

    #Menünün birinci ve ikinci bölümlerinden pizza ve sos seçeneklerini attım.
    t_options = menu_parts[3].strip()
    print("\n"+t_options)     

    # Kullanıcının tercihlerine göre pizza ve sos nesneleri oluşturdum.
    pizza = PIZZA[pizza_secimi]()
    sos = SOS[sos_secimi](pizza)

    # Toplam maliyeti hesapladım.
    total = sos.get_cost()

    # Kullanıcının kişisel ve kredi kartı bilgilerini istedim.
    # Ad Doğrulaması 
    while True:
            ad = input("\nLütfen adınızı giriniz: ")
            if re.match("^[a-zA-Z ]+$", ad):
               break
            else:
               print("Geçersiz isim. Lütfen yalnızca alfabetik karakterler ve boşluklar girin.")

    # TC Kimlik Doğrulaması
    while True:
            tr_id = input("\nLütfen TC kimlik numaranızı giriniz: ")
            if len(tr_id) == 11 and tr_id.isnumeric():
              break
            else:
             print("Geçersiz TC kimlik numarası. Lütfen 11 haneli geçerli bir sayı girin.")

    # Kredi Kartı Numarası Doğrulama
    while True:
            kredi_kart_numarasi = input("\nLütfen kredi kartı numaranızı girin: ")
            if re.match("^([0-9]{4}-){3}[0-9]{4}|[0-9]{16}$", kredi_kart_numarasi): # Geçerli format olarak kullandığım maske.
               break
            else:
               print("Geçersiz kredi kartı numarası. Geçerli bir kredi kartı numarası girin.")

    # Kredi Kartı Şifre Doğrulama
    while True:
            kredi_kart_sifresi = input("\nLütfen kredi kartı şifrenizi giriniz: ")
            if len(kredi_kart_sifresi) == 4 and kredi_kart_sifresi.isnumeric():
               break
            else:
              print("Geçersiz kredi kartı şifresi. Lütfen 4 haneli geçerli bir sayı girin.")
    
    

    # Sipariş zamanı için bir zaman damgası oluşturdum.
    siparis_zamani = datetime.datetime.now()

    # Sipariş bilgilerini veritabanı dosyasına yazdım.
    with open("Orders_Database.csv", "a") as file:
        writer = csv.writer(file)
        writer.writerow([ad, tr_id, kredi_kart_numarasi, sos.get_description(), siparis_zamani, kredi_kart_sifresi])

    # Kullanıcıya bir onay mesajı yazdırdım.
    print(f"Teşekkürler, {ad}! Sipariş verdiğin {sos.get_description()} ile birlikte hazırlandı. Toplam ödediğin ücret {total:.2f} TL.")

main()
```
